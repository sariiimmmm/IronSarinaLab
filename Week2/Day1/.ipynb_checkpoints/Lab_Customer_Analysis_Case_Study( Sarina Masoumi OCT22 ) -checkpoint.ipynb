```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import boxcox
pd.options.display.max_rows = 100
```

```python
def load_original_data():
    file1 = pd.read_csv('Data/file1.csv')
    file2 = pd.read_csv('Data/file2.csv')
    file3 = pd.read_csv('Data/file3.csv')
    return pd.concat([file1,file2,file3], axis=0)
```

```python
auto_insurance=load_original_data()
auto_insurance=auto_insurance.reset_index()
auto_insurance
```

```python
def lower_case_columns(x):
    x.columns=[i.lower() for i in x.columns]
    return x
```

```python
auto_insurance["Gender"]=auto_insurance['Gender'].fillna(auto_insurance["GENDER"])
auto_insurance["State"]=auto_insurance['State'].fillna(auto_insurance["ST"])
auto_insurance
```

```python
auto_insurance.drop_duplicates()
```

```python
def drop_columns(x,a) :
    x.drop(columns=[a], inplace=True)
    return x
```

```python
drop_columns(auto_insurance,"GENDER")
drop_columns(auto_insurance,"ST")
```

```python
def clean_gender(x):
    if x in ['M', 'MALE',"male"]:
        return 'Male'
    elif x in ['F', 'female',"male"]:
        return 'Female'
    elif np.nan:  pass
    else:
        return 'U'
```

```python
auto_insurance['gender'] = list(map(clean_gender, auto_insurance['gender'])) 
auto_insurance
```

```python
def clean_city(x):
    if x in ['WA', "Washington"]:
        return "Washington"
    elif x in ["Arizona","AZ"]:
        return "Arizona"
    elif x in ["California","Cali"]:
        return "California"
    elif np.nan:  pass
    else:
        return 'U'
```

```python
auto_insurance["state"] = list(map(clean_city, auto_insurance["state"])) 
auto_insurance
```
